```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
def plot_link(ax_1, node_1, node_2, counts, count_0=0,color='blue'):
    z_1 = np.linspace(count_0,counts,100)
    x_1 = z_1 * 0 + node_1
    y_1 = z_1 * 0 + node_2
    ax_1.plot(x_1, y_1, z_1, color=color)
```

```python
def plot_dash_lines(ax_1, node_1, node_2, x_lim, y_lim):
    x_1 = np.linspace(node_1, node_1, 100)
    y_1= np.linspace(node_2, y_lim, 100)
    ax_1.plot(x_1,y_1,zs=0, lw=1, linestyle='--', color='black')

    x_2 = np.linspace(x_lim, node_1,100)
    y_2= np.linspace(node_2, node_2,100)
    ax_1.plot(x_2,y_2,zs=0, lw=1, linestyle='--', color='black')

```

```python
fig=plt.figure()
ax = fig.gca(projection='3d')
ax.xaxis.set_pane_color((1,1,1,0))
ax.xaxis._axinfo['grid']['color'] = (1,1,1,0)
ax.yaxis.set_pane_color((1,1,1,0))
ax.yaxis._axinfo['grid']['color'] = (1,1,1,0)
ax.zaxis.set_pane_color((1,1,1,0))
ax.zaxis._axinfo['grid']['color'] = (1,1,1,0)

theta_1 = 7.
theta_2 = 3.
theta_3 = 1.
min_lim = 0.9
max_lim = 8.1


edges = [(theta_3, theta_3, 4),
         (theta_2, theta_3, 2),
         (theta_2, theta_1, 1),
         (theta_1, theta_2, 3),
         (theta_1, theta_3, 1)]

for edge in edges:
    i, j, c = edge
    plot_link(ax, i, j, c)
    plot_dash_lines(ax, i, j, max_lim, max_lim)





ticks = [theta_1, theta_2, theta_3]
ticklabels = [r'$\theta_' + str(i+1) + r'$' for i in range(len(ticks))]
ax.set_xticks(ticks)
ax.set_xticklabels(ticklabels)
ax.set_yticks(ticks)
ax.set_yticklabels(ticklabels)
ax.set_xlim(min_lim, max_lim)
ax.set_ylim(min_lim, max_lim)
ax.set_zlabel('Counts')
ax.view_init(elev=30, azim=60)
plt.tight_layout(pad=0.05)
plt.savefig("output3.jpg", bbox_inches='tight', pad_inches=0.05)
plt.show()
```
